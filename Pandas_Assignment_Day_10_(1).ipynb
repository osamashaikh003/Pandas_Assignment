HINT -
- Define a function c_to_f which takes input temp in celsius and returns a value with temperature in Fahrenheit.
- To check: c_to_f(10) should return 50.


```python
# write function here

def c_to_f(x):
  temp_f=(x*9/5)+32
  return temp_f



```

## Groupby

```python
# Create a dataframe using dictionary of your choice

data1 =pd.DataFrame({'Department':['HR','IT','Sales','HR','IT','Sales'],
                    'Salary':[50000,60000,70000,45000,75000,80000]})




```

```python
# Use Groupby of single column with aggregate sum()
grouped=data1.groupby('Department')['Salary'].sum()

print(grouped)
```

```python
# Use Groupby of single column with aggregate count()

counted=data1.groupby('Department')['Salary'].count()

print(counted)
```

```python
# Use Groupby of single column with aggregate min() and max()

min_max=data1.groupby('Department')['Salary'].agg(['min','max'])
print(min_max)
```

```python
# Use Groupby of any 2 columns with aggregate mean()

mean=data1.groupby(['Department','Salary'])['Salary'].mean()
print(mean)
```

```python
# Use Groupby of any 2 columns with aggregate min() and max()

min_max1=data1.groupby(['Department','Salary'])['Salary'].agg(['min','max'])
print(min_max1)
```
